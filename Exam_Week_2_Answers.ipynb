## Problem 3: Fill the database
Using pymysql, create a Python script that reads the data from these files, and fills the database.  There are a variety of strategies to accomplish this.  I will give all strategies equal credit - do whichever one you are most confident with.

```python
import csv
import pymysql.cursors

Germplasm_Exam_2 = open("Germplasm.tsv", "r")
Dict_germplasm = csv.DictReader(Germplasm_Exam_2, delimiter = "\t", quotechar = '"')
#Germplasm_Exam_2.seek(0)
#Germplasm_Exam_2.readline()



LocusGene = open("LocusGene.tsv", "r")
Dict_locusgene = csv.DictReader(LocusGene, delimiter = "\t", quotechar = '"')
#LocusGene.seek(0)
#LocusGene.readline()


#for row in Dict_germplasm:
    #Germplasm_Exam_2.readline()
    #print("Locus_ID: {}".format(row["Locus"]))

#establish a connection with the database.
connection = pymysql.connect(host='localhost',
                             user='root',
                             password='root',
                             db='germplasm',
                             charset='utf8mb4',  
                             cursorclass=pymysql.cursors.DictCursor)
#connection.autocommit = True

try:
    with connection.cursor() as cursor:
        
        for row in Dict_germplasm:
            
            sql = """INSERT INTO germplasm_tsv_3 (Locus, germplasm, phenotype, pubmed)
            VALUES ('{}', '{}', '{}', '{}');""".format(row["Locus"], row["germplasm"], row["phenotype"], row["pubmed"])
            cursor.execute(sql)
        
        for row in Dict_locusgene:
            
            sql = """INSERT INTO locusgene_tsv_3 (Locus_2, gene, protein_length)
            VALUES ('{}', '{}', '{}');""".format(row["Locus"], row["Gene"], row["ProteinLength"])
            cursor.execute(sql) 
    connection.commit()

finally:
    connection.close()
    Germplasm_Exam_2.close()
    LocusGene.close()
        
```
